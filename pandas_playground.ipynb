Importing modeules

```python
import numpy as np
import pandas as pd
```

Reading the csv file

```python
# LOAD THE DATA INTO A PANDAS DATA FRAME CALLED: df_raw
df_raw = pd.read_csv('GlobalLandTemperaturesByCity.csv')
```

Modifying the file

```python
### BEGIN SOLUTION
df=df_raw.copy()
df['dt'] = pd.to_datetime(df['dt'])
df_drop = df.dropna(subset = ['AverageTemperature'])
df_time=df_drop.copy()
df_time = df_time.set_index('dt')
### END SOLUTION
```

Creating a sub dataframe

```python
### BEGIN SOLUTION
count_Yemen=df_time['AverageTemperature'][(df_time['Country'] == 'Yemen')].count()
print(count_Yemen)
df_Yemen=df_time['AverageTemperature'][(df_time['City'] == 'Aden') & (df_time['Country'] == 'Yemen')]
df_Yemen_Mean=df_Yemen.resample('Y').mean()
df_Yemen_Mean.plot()
### END SOLUTION
```
